```python
df["Time_slot"]=0
j=0
for i in df["req_hour"]:
    if df.iloc[j,6]<5:
        df.iloc [j,8]="Pre_Morning"
    elif 5<=df.iloc[j,6]<10:   
         df.iloc [j,8]="Morning_Rush"
            
    elif 10<=df.iloc[j,6]<17:
         df.iloc [j,8]="Day_Time"
            
    elif 17<=df.iloc[j,6]<22:
         df.iloc [j,8]="Evening_Rush"
    else:
         df.iloc[j,8]="Late_Night"
    j =j+1
```
